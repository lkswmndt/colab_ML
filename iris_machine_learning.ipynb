```python
features = iris_data.iloc[:,0:4]
label = iris_data.iloc[:,4]
print('Features: ')
print(features)
print('Labels: ')
print(label)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    features,
    label,
    test_size= 0.2,
    random_state=0
    )
print('Jumlah Data Train:', X_train.shape[0])
print('Jumlah Data Test:', X_test.shape[0])
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5, weights='uniform')

knn.fit(X_train, Y_train)
train_acc = knn.score(X_train, Y_train)

print('The Accuracy of KNN Classifier on training data is: {:.3f}'.format(train_acc))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3, weights='distance')

knn.fit(X_train, Y_train)
train_acc = knn.score(X_train, Y_train)

print('The Accuracy of KNN Classifier on training data is: {:.3f}'.format(train_acc))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=8, weights='distance')

knn.fit(X_train, Y_train)
train_acc = knn.score(X_train, Y_train)

print('The Accuracy of KNN Classifier on training data is: {:.3f}'.format(train_acc))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=10, weights='distance')

knn.fit(X_train, Y_train)
train_acc = knn.score(X_train, Y_train)

print('The Accuracy of KNN Classifier on training data is: {:.3f}'.format(train_acc))
```

```python
knn_test_acc = knn.score(X_test, Y_test)

print('The Accuracy of KNN Classifier on training data is: {:.3f}'.format(knn_test_acc))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay.from_estimator(
    knn,
    X_test,
    Y_test
    )
disp.ax_.set_title('Confusion Matrix')
```

# PARAMETER TUNING

```python
from subprocess import check_output
from sklearn import tree
from sklearn.pipeline import make_pipeline
from sklearn import preprocessing
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv('iris.data')
features = data.iloc[:,0:4]
label = data.iloc[:,4]
```

```python
from sklearn.neighbors import KNeighborsClassifier

#mengatur nilai parameter n_range dalam sebuah list yang berisikan angka 1-10
n_range = list(range(1,10))

#mengatur nilai
weight_list = ['uniform', 'distance']

param_grid = {
    'n_neighbors': n_range,
    'weights': weight_list
    }

grid = GridSearchCV(
    estimator= KNeighborsClassifier(),
    param_grid= param_grid,
    cv= 3,
    scoring= 'accuracy',
    refit= True
    )


knn_model = make_pipeline(preprocessing.StandardScaler(), grid)

knn_model.fit(features, label)

print('Accuracy of the tuned model:', round(grid.best_score_, 4))
print('Best parameter', grid.best_params_)
```

```python
import pickle

with open('knn_model.pickle', 'wb') as f:
  pickle.dump(knn_model, f)
```

# MAKING PREDICTION

```python
with open('knn_model.pickle', 'rb') as f:
  model = pickle.load(f)

new_data = [6.5, 3.5, 5.5, 2.0]
X= np.array([new_data])
prediction = model.predict(X)
print('Prediction of species: {}', format(prediction))
```
